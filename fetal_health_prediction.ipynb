# Weak Stacking Classifier

```python
from sklearn.ensemble import StackingClassifier

# Define weak base models
weak_base_models = [
    ('lr', LogisticRegression(max_iter=1000, random_state=42)),
    ('nb', GaussianNB()),
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('ab', AdaBoostClassifier(n_estimators=100, random_state=42))
]

# Meta-model (final estimator)
weak_meta_model = SVC(probability=True, random_state=42)

# Create Weak Stacking Classifier
weak_stacking_clf = StackingClassifier(estimators=weak_base_models, final_estimator=weak_meta_model, cv=5)

# Train the model
weak_stacking_clf.fit(X_train, y_train)

# Predictions
y_pred_weak_stack = weak_stacking_clf.predict(X_test)

# Evaluation
print("\n Weak Stacking Classifier Metrics")
print("Accuracy:", accuracy_score(y_test, y_pred_weak_stack))
print(classification_report(y_test, y_pred_weak_stack))

# Confusion Matrix
plt.figure(figsize=(6,5))
sns.heatmap(confusion_matrix(y_test, y_pred_weak_stack), annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - Weak Stacking Classifier")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

# Strong Stacking Classifier

```python
# Define strong base models with optimized hyperparameters
strong_base_models = [
    ('rf', RandomForestClassifier(n_estimators=100, max_depth=8, min_samples_split=10, random_state=42)),
    ('gb', GradientBoostingClassifier(n_estimators=100, learning_rate=0.07, max_depth=4, random_state=42)),
    ('svm', SVC(probability=True, kernel='rbf', C=3.0, gamma='scale', random_state=42))
]

# Meta-model for strong stacking (Optimized LightGBM)
strong_meta_model = lgb.LGBMClassifier(n_estimators=150, learning_rate=0.05, max_depth=7, random_state=42)

# Create Strong Stacking Classifier
strong_stacking_clf = StackingClassifier(estimators=strong_base_models, final_estimator=strong_meta_model, cv=5)

# Train the model
strong_stacking_clf.fit(X_train, y_train)

# Predictions
y_pred_strong_stack = strong_stacking_clf.predict(X_test)

# Evaluation
print("\n🔹 Strong Stacking Classifier Metrics")
print("Accuracy:", accuracy_score(y_test, y_pred_strong_stack))
print(classification_report(y_test, y_pred_strong_stack))

# Confusion Matrix
plt.figure(figsize=(6,5))
sns.heatmap(confusion_matrix(y_test, y_pred_strong_stack), annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - Strong Stacking Classifier")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```
